```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set(palette='rainbow')
```

```python
import scipy.stats as stats
np.set_printoptions(precision=2,suppress=True)
```

```python
data = pd.read_csv('budget_revenue_2017.csv',parse_dates=['release_date'],index_col='release_date')
```

```python
data.head()
```

```python
##getting only data for 2016
```

```python
data_16 = data.loc['2016']
```

```python
data_16
```

```python
#### visualize data
sns.lmplot(data=data_16,x='budget',y='revenue',height=8,ci=None)
plt.show()
```

```python
#linear regression with numpy
```

```python
x = data_16.iloc[:,-2].values #budget
```

```python
y = data_16.iloc[:,-1].values #revenue
```

```python
regression_m = np.polyfit(x=x,y=y,deg=1)
regression_m
```

```python
np.polyfit(x=data_16['budget'],y=data_16['revenue'],deg=1)
```

```python
### creating numpy x and y values for regression
```

```python
X1 = np.array([min(x),max(x)])
```

```python
X1
```

```python
Y1 = np.polyval(regression_m,X1)
```

```python
Y1
```
